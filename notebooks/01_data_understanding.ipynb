# Customer Churn Prediction: Profit-Based Evaluation of ML Models
## Step 1: Data Collection & Understanding

```python
# ── IMPORTS ──────────────────────────────────────────────────────────────
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.2f}'.format)
sns.set_theme(style='whitegrid', palette='muted')

print('Libraries loaded successfully ✓')
```

```python
# ── LOAD DATASET ─────────────────────────────────────────────────────────
# Update the path below if your CSV is in a different folder
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')

print(f'Dataset loaded successfully ✓')
print(f'Shape: {df.shape[0]} rows × {df.shape[1]} columns')
```

```python
# ── FIRST LOOK ───────────────────────────────────────────────────────────
print('=== First 5 Rows ===')
df.head()
```

```python
# ── COLUMN NAMES & DATA TYPES ────────────────────────────────────────────
print('=== Column Names and Data Types ===')
print(df.dtypes)
```

```python
# ── STATISTICAL SUMMARY ──────────────────────────────────────────────────
print('=== Statistical Summary (Numerical Columns) ===')
df.describe()
```

```python
# ── CATEGORICAL SUMMARY ──────────────────────────────────────────────────
print('=== Categorical Columns Summary ===')
cat_cols = df.select_dtypes(include='object').columns.tolist()
print(f'Categorical columns ({len(cat_cols)}): {cat_cols}\n')

for col in cat_cols:
    print(f'{col}: {df[col].nunique()} unique values → {df[col].unique()[:5]}')
```

```python
# ── MISSING VALUES CHECK ─────────────────────────────────────────────────
print('=== Missing Values ===')
missing = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df)) * 100
missing_df = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_pct})
print(missing_df[missing_df['Missing Count'] > 0])

if missing_df['Missing Count'].sum() == 0:
    print('No missing values found at surface level ✓')
    print('Note: TotalCharges may have hidden spaces — will fix in Step 2.')
```

```python
# ── CHURN DISTRIBUTION ───────────────────────────────────────────────────
print('=== Target Variable: Churn Distribution ===')
churn_counts = df['Churn'].value_counts()
churn_pct    = df['Churn'].value_counts(normalize=True) * 100

print(churn_counts)
print(f'\nChurn Rate: {churn_pct["Yes"]:.1f}%')
print(f'Retention Rate: {churn_pct["No"]:.1f}%')

# Plot
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Bar chart
churn_counts.plot(kind='bar', ax=axes[0], color=['steelblue', 'tomato'], edgecolor='white', rot=0)
axes[0].set_title('Churn Count', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Churn')
axes[0].set_ylabel('Number of Customers')
for p in axes[0].patches:
    axes[0].annotate(f'{int(p.get_height()):,}', (p.get_x() + p.get_width()/2, p.get_height()),
                     ha='center', va='bottom', fontweight='bold')

# Pie chart
axes[1].pie(churn_counts, labels=['No Churn', 'Churn'], autopct='%1.1f%%',
            colors=['steelblue', 'tomato'], startangle=90,
            wedgeprops={'edgecolor': 'white', 'linewidth': 2})
axes[1].set_title('Churn Proportion', fontsize=13, fontweight='bold')

plt.suptitle('Target Variable Distribution', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('plot_01_churn_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
print('Plot saved ✓')
```

```python
# ── KEY NUMERIC FEATURES OVERVIEW ────────────────────────────────────────
# TotalCharges is stored as object — preview the issue before fixing in Step 2
print('=== TotalCharges column (object type — needs fixing) ===')
print(df['TotalCharges'].head(10))
print(f'\nSample of blank/space values: {(df["TotalCharges"].str.strip() == "").sum()} found')

# Tenure and MonthlyCharges look
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

sns.histplot(df['tenure'], bins=30, ax=axes[0], color='steelblue', kde=True)
axes[0].set_title('Customer Tenure Distribution', fontweight='bold')
axes[0].set_xlabel('Tenure (months)')

sns.histplot(df['MonthlyCharges'], bins=30, ax=axes[1], color='darkorange', kde=True)
axes[1].set_title('Monthly Charges Distribution', fontweight='bold')
axes[1].set_xlabel('Monthly Charges ($)')

plt.suptitle('Key Numeric Features Overview', fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('plot_02_numeric_overview.png', dpi=150, bbox_inches='tight')
plt.show()
print('Plot saved ✓')
```
